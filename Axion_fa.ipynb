# Constraints on $1/f_a$ scale versus axion mass

QCD axion band defined by
\begin{equation}
m_{a}=5.70(7) \mu \mathrm{eV}\left(\frac{10^{12} \,\mathrm{GeV}}{f_{a}}\right)
\end{equation}

```python
from numpy import *
import matplotlib.pyplot as plt
from PlotFuncs import BlackHoleSpins,Axion_fa,MySaveFig
    
fig,ax = Axion_fa.FigSetup()

BlackHoleSpins(ax,1,[0.6e-13,0.4e-17],fs=27,rotation=90,text_col='k',alpha=0.8,facecolor='gray')
Axion_fa.QCDAxion(ax)
Axion_fa.nEDM(ax)
Axion_fa.SolarCore(ax)
Axion_fa.Pulsars(ax)
Axion_fa.SN1987A(ax)

plt.gcf().text(0.47,0.43,'Stellar',ha='center',alpha=0.75)
plt.gcf().text(0.47,0.39,'BHs',ha='center',alpha=0.75)
plt.gcf().text(0.197,0.18,'SM \n BHs',alpha=0.75,ha='center')

MySaveFig(fig,'Axion_fa')
```

```python
from numpy import *
import matplotlib.pyplot as plt
from PlotFuncs import BlackHoleSpins,Axion_fa,MySaveFig
    

def NeutronStars(ax,text_pos=[0.5e-4,1.1e-13],col='#1f6ff0',text_col='#1f6ff0',text_rot=43.5,fs=29,zorder=-5):
    dat = loadtxt('limit_data/fa/Projections/NeutronStars.txt')
    plt.plot(dat[:,0],dat[:,1],'--',color=col,lw=3,alpha=1,zorder=zorder)
    plt.fill_between(dat[:,0],dat[:,1],y2=1e0,color=col,zorder=zorder,alpha=0.1)
    plt.text(text_pos[0],text_pos[1],r'{\bf Neutron stars}',color=text_col,rotation=text_rot,fontsize=fs,ha='right')
    return

def Inspirals(ax,text_pos=[1e-16,2e-14],col='purple',text_col='purple',text_rot=0,fs=23,zorder=-2):
    dat = loadtxt('limit_data/fa/Projections/NSBH-Inspiral.txt')
    plt.plot(dat[:,0],dat[:,1],'--',color=col,lw=3,alpha=1,zorder=zorder)
    plt.fill_between(dat[:,0],dat[:,1],y2=1e-99,color=col,zorder=zorder,alpha=0.1)
    
    dat = loadtxt('limit_data/fa/Projections/NSNS-Inspiral.txt')
    plt.plot(dat[:,0],dat[:,1],'--',color=col,lw=3,alpha=1,zorder=zorder)
    plt.fill_between(dat[:,0],dat[:,1],y2=1e-99,color=col,zorder=zorder,alpha=0.1)
    
    plt.text(text_pos[0],text_pos[1],r'{\bf Inspirals}',color=text_col,rotation=text_rot,fontsize=fs)
    return
    
fig,ax = Axion_fa.FigSetup()

BlackHoleSpins(ax,1,[0.6e-13,0.4e-17],text_on=False)
Axion_fa.QCDAxion(ax)
Axion_fa.nEDM(ax)
Axion_fa.SolarCore(ax,text_pos=[0.1e-10,0.12e-10])
Axion_fa.Pulsars(ax)
Axion_fa.SN1987A(ax)
Inspirals(ax)
NeutronStars(ax)

plt.gcf().text(0.47,0.43,'Stellar',ha='center',alpha=0.75)
plt.gcf().text(0.47,0.39,'BHs',ha='center',alpha=0.75)
plt.gcf().text(0.197,0.18,'SM \n BHs',alpha=0.75,ha='center')

MySaveFig(fig,'Axion_fa_with_Projections')
```
