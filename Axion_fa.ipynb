# Constraints on $1/f_a$ scale versus axion mass

QCD axion band defined by
\begin{equation}
m_{a}=5.70(7) \mu \mathrm{eV}\left(\frac{10^{12} \,\mathrm{GeV}}{f_{a}}\right)
\end{equation}

```python
from numpy import *
import matplotlib.pyplot as plt
from PlotFuncs import BlackHoleSpins,Axion_fa,MySaveFig,FigSetup
    
fig,ax = FigSetup(Shape='Rectangular',ylab=r'$1/f_a$ [GeV$^{-1}$]',mathpazo=True,\
                              g_min=1e-20,g_max=1e-7,m_min=1e-20,m_max=1e-2)

BlackHoleSpins(ax,1,[0.6e-13,0.4e-17],fs=27,rotation=90,text_col='k',alpha=1,facecolor='gray')
Axion_fa.QCDAxion(ax)
Axion_fa.nEDM(ax)
Axion_fa.SolarCore(ax)
Axion_fa.Pulsars(ax)
Axion_fa.SN1987A(ax)


plt.gcf().text(0.47,0.43,'Stellar',ha='center',alpha=0.75)
plt.gcf().text(0.47,0.39,'BHs',ha='center',alpha=0.75)
plt.gcf().text(0.197,0.18,'SM \n BHs',alpha=0.75,ha='center')

MySaveFig(fig,'Axion_fa')
```
